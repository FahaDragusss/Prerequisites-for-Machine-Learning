```python
import pandas as pd
import numpy as np
```

```python
# Create Data
stuff = {'Corporations':['Apple','Google','Meta','Apple','Google','Meta']
         ,'Employees':['Marie','Zagreus','Melissa','Fedosev','Saun','Jokeph']
         ,'Salary':[200,220,190,130,120,150]}
# Create Dataframe
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Sort Salaries from lowest to highest
my_df.sort_values('Salary')
```

```python
# Sort Salaries from highest to lowest
my_df.sort_values('Salary', ascending=False)
```

```python
# Inplace 
my_df.sort_values('Salary', ascending=False, inplace=True)
my_df
```

```python
# Sort strings Alphabetically 
my_df.sort_values('Corporations')
```
