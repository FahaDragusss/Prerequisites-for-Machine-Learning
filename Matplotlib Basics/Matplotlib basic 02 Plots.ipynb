```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
%matplotlib inline

# Create a DataFrame
my_df = pd.DataFrame(randn(25, 4), columns=['A', 'B', 'C', 'D'])
my_df
```

```python
# Area plot
my_df.abs().plot(kind='area',)
```

```python
# Shading
my_df.abs().plot(kind='area',alpha=0.5)
```

```python
# Title
my_df.abs().plot(kind='area',alpha=0.5,title='Area Plot')
```

```python
# Plot only 1 column
my_df['A'].abs().plot(kind='area',alpha=0.5,title='Area Plot')
```

```python
# Plot more than 1 column
my_df[['A','B']].abs().plot(kind='area',alpha=0.5,title='Area Plot')
```
