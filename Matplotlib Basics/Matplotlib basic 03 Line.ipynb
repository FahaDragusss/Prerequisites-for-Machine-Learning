```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(24,4), columns=['A','B','C','D'])
my_df
```

```python
# Line plot
my_df.abs().plot(kind='line')
```

```python
# Line width
my_df.abs().plot(kind='line', lw = 3)
```

```python
# Line for specific columns
my_df[['A','B']].abs().plot(kind='line', lw = 3)
```

```python
# Size
my_df[['A','B']].abs().plot(kind='line', lw = 3, figsize=(10,5))
```

```python
# Title
my_df[['A','B']].abs().plot(kind='line', lw = 3, figsize=(10,5), title='My Line Plot')
```

```python
# Shading
my_df[['A','B']].abs().plot(kind='line', lw = 3, figsize=(10,5), title='My Line Plot', alpha=0.5) 
```

```python
# Legend
my_df[['A','B']].abs().plot(kind='line', lw = 3, figsize=(10,5), title='My Line Plot', alpha=0.5, legend=True)
```
