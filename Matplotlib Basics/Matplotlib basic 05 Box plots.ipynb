```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(500,4), columns = ['A', 'B', 'C', 'D'])
my_df
```

```python
# Box plots
my_df.plot.box()
```

```python
# Boxplots size
my_df.plot.box(figsize=(10,6))
```

```python
# Boxplots color
my_df.plot(kind='box', title = " My Boxplot", color = {'medians': 'r', 'boxes': 'b', 'whiskers': 'g', 'caps': 'black'})
```
