```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn

my_df = pd.DataFrame(randn(500, 4), columns=['A', 'B','C','D'])
my_df
```

```python
# Hex bin
my_df.plot(kind='hexbin', x = 'A',y = 'B')
```

```python
# Grid size
my_df.plot(kind='hexbin', x = 'A',y = 'B', gridsize=25)
```
